```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set starting value at -100, 100.5 (to include 100), with a step of 0.5
x = np.arange(-100, 100.5,0.5)
```

```python
# Y is an array that is the cos of
y = np.cos(x)
```

```python
# Z is an array that is the sine of x
z = np.sin(x)
```

```python
# Editd length and width to show data better
plt.figure(figsize=(15,10))

# Legend titles
plt.plot(x, y, label='Cosine of y', color='red')
plt.plot(x, z, label='Sine of z', color='orange')
plt.legend()

# Axis labels
plt.xlabel('X value')
plt.ylabel('Y and Z values')

# Show plot
plt.show()
```
